```python
# Exercício 1 — Formação de times a partir de amizades

from collections import defaultdict

def dfs(grafo, vertice, visitados):
    visitados.add(vertice)

    for vizinho in grafo[vertice]:
        if vizinho not in visitados:
            dfs(grafo, vizinho, visitados)


def contar_grupos(N, amizades):
    grafo = defaultdict(list)

    # Construção do grafo não direcionado
    for u, v in amizades:
        grafo[u].append(v)
        grafo[v].append(u)

    visitados = set()
    grupos = 0

    # Conta componentes conectadas
    for aluno in range(1, N + 1):
        if aluno not in visitados:
            dfs(grafo, aluno, visitados)
            grupos += 1

    return grupos


# Exemplo
N = 6
amizades = [
    (1, 2),
    (2, 3),
    (4, 5)
]

print(contar_grupos(N, amizades))
```

Cada aluno é um vértice do grafo e cada amizade é uma aresta não direcionada.

A DFS percorre todos os alunos conectados direta ou indiretamente. Cada nova DFS iniciada representa um novo grupo.

Complexidade
Construção do grafo: O(M)
DFS completa: O(N + M)

Complexidade total:

O(N + M)

```python
# Exercício 2 — Validação de passeios em uma rede de túneis

from collections import defaultdict, deque

def bfs(grafo, origem, destino):
    fila = deque([origem])
    visitados = set([origem])

    while fila:
        atual = fila.popleft()

        if atual == destino:
            return True

        for vizinho in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)

    return False


def contar_passeios_validos(S, tuneis, passeios):
    grafo = defaultdict(list)

    # Grafo não direcionado
    for x, y in tuneis:
        grafo[x].append(y)
        grafo[y].append(x)

    validos = 0

    for passeio in passeios:
        ok = True

        for i in range(len(passeio) - 1):
            origem = passeio[i]
            destino = passeio[i + 1]

            if not bfs(grafo, origem, destino):
                ok = False
                break

        if ok:
            validos += 1

    return validos


# Exemplo
S = 5

tuneis = [
    (1, 2),
    (2, 3),
    (3, 4)
]

passeios = [
    [1, 3, 4],
    [1, 5],
    [2, 4]
]

print(contar_passeios_validos(S, tuneis, passeios))
```

Complexidade

Para cada verificação usamos BFS:

O(S + T)

Se houver P passeios e K pares consecutivos totais:

O(K * (S + T))

```python
# Exercício 3 — Conectividade dinâmica entre ilhas

from collections import defaultdict, deque

def bfs(grafo, origem, destino):
    fila = deque([origem])
    visitados = set([origem])

    while fila:
        atual = fila.popleft()

        if atual == destino:
            return True

        for vizinho in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)

    return False


def processar_operacoes(N, operacoes):
    grafo = defaultdict(list)

    for tipo, a, b in operacoes:

        # Construção de ponte
        if tipo == 1:
            grafo[a].append(b)
            grafo[b].append(a)

        # Consulta
        else:
            print(1 if bfs(grafo, a, b) else 0)


# Exemplo
operacoes = [
    (1, 1, 2),
    (1, 2, 3),
    (0, 1, 3),
    (0, 1, 4)
]

processar_operacoes(4, operacoes)
```

Comparação DFS × BFS
DFS
Explora profundamente um caminho antes de voltar
Pode ir rapidamente para regiões distantes do grafo
BFS
Explora por níveis
Visita primeiro os vértices mais próximos da origem
Complexidade
O(V + E)

```python
# Exercício 6 — Menor caminho em rede social

from collections import deque

grafo = {
    "Idris": ["Kamil", "Talia"],
    "Kamil": ["Idris", "Lina"],
    "Lina": ["Kamil", "Sasha"],
    "Sasha": ["Lina", "Marco"],
    "Marco": ["Sasha", "Ken"],
    "Ken": ["Marco", "Talia"],
    "Talia": ["Idris", "Ken"]
}

def bfs_menor_caminho(origem, destino):
    fila = deque([origem])

    visitados = set([origem])

    pai = {
        origem: None
    }

    while fila:
        atual = fila.popleft()

        if atual == destino:
            break

        for vizinho in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                pai[vizinho] = atual
                fila.append(vizinho)

    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = pai[atual]

    caminho.reverse()

    return caminho, len(caminho) - 1


caminho, distancia = bfs_menor_caminho("Idris", "Lina")

print("Caminho:", caminho)
print("Distância:", distancia)
```

Saída esperada
Caminho: ['Idris', 'Kamil', 'Lina']
Distância: 2
Por que é mínimo?

BFS visita os vértices em camadas. O primeiro caminho encontrado possui o menor número de arestas.

```python
# Exercício 7 — Travessia em grafo direcionado

from collections import deque

grafo = {
    "Inicio": ["A", "B"],
    "A": ["C"],
    "B": ["C", "F"],
    "C": ["D"],
    "D": ["E"],
    "E": [],
    "F": ["E"]
}

# DFS
visitados = set()
ordem_dfs = []

def dfs(v):
    visitados.add(v)
    ordem_dfs.append(v)

    for vizinho in grafo[v]:
        if vizinho not in visitados:
            dfs(vizinho)

dfs("Inicio")

print("DFS:", ordem_dfs)

# BFS
fila = deque(["Inicio"])
visitados = set(["Inicio"])
ordem_bfs = []

while fila:
    atual = fila.popleft()
    ordem_bfs.append(atual)

    for vizinho in grafo[atual]:
        if vizinho not in visitados:
            visitados.add(vizinho)
            fila.append(vizinho)

print("BFS:", ordem_bfs)
```

Comparação
DFS

Aprofunda caminhos antes de retornar.

BFS

Explora vértices por níveis.

Influência das arestas direcionadas

Só é possível seguir o sentido definido pelas arestas.

Complexidade

```python
# Exercício 8 — Viabilidade operacional na movimentação de containers

from collections import defaultdict, deque

grafo = defaultdict(list)

arestas = [
    ("Berco_A", "Patio_1", 4),
    ("Berco_A", "Patio_2", 7),
    ("Berco_B", "Patio_2", 3),
    ("Berco_B", "Patio_3", 6),
    ("Patio_1", "Patio_2", 2),
    ("Patio_2", "Patio_3", 2),
    ("Patio_1", "Alfandega", 8),
    ("Patio_2", "Alfandega", 5),
    ("Patio_3", "Centro_Logistico", 4),
    ("Alfandega", "Centro_Logistico", 3)
]

for u, v, peso in arestas:
    grafo[u].append((v, peso))
    grafo[v].append((u, peso))

def bfs(origem, destino):
    fila = deque([origem])

    visitados = set([origem])

    pai = {
        origem: None
    }

    while fila:
        atual = fila.popleft()

        if atual == destino:
            break

        for vizinho, _ in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                pai[vizinho] = atual
                fila.append(vizinho)

    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = pai[atual]

    caminho.reverse()

    return caminho

caminho = bfs("Berco_A", "Centro_Logistico")

print(caminho)
```

Explicação

BFS ignora os pesos e minimiza apenas o número de arestas.

O caminho encontrado pode não possuir o menor custo operacional.

```python
# Exercício 9 — Menor custo com Dijkstra

import heapq
from collections import defaultdict

grafo = defaultdict(list)

arestas = [
    ("Berco_A", "Patio_1", 4),
    ("Berco_A", "Patio_2", 7),
    ("Berco_B", "Patio_2", 3),
    ("Berco_B", "Patio_3", 6),
    ("Patio_1", "Patio_2", 2),
    ("Patio_2", "Patio_3", 2),
    ("Patio_1", "Alfandega", 8),
    ("Patio_2", "Alfandega", 5),
    ("Patio_3", "Centro_Logistico", 4),
    ("Alfandega", "Centro_Logistico", 3)
]

for u, v, peso in arestas:
    grafo[u].append((v, peso))
    grafo[v].append((u, peso))

def dijkstra(origem):
    dist = {v: float("inf") for v in grafo}
    dist[origem] = 0

    pai = {origem: None}

    heap = [(0, origem)]

    while heap:
        custo, atual = heapq.heappop(heap)

        if custo > dist[atual]:
            continue

        for vizinho, peso in grafo[atual]:
            novo = custo + peso

            if novo < dist[vizinho]:
                dist[vizinho] = novo
                pai[vizinho] = atual
                heapq.heappush(heap, (novo, vizinho))

    return dist, pai

distancias, pais = dijkstra("Berco_A")

print(distancias)
```

Complexidade

Usando heap:

O((V + E) log V)

```python
# Exercício 10 — Reconstrução de rota ótima

def reconstruir_caminho(pai, destino):
    caminho = []

    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = pai[atual]

    caminho.reverse()

    return caminho


caminho = reconstruir_caminho(
    pais,
    "Centro_Logistico"
)

print(caminho)
```

Exercício 11 — Impacto dos pesos na escolha de rotas
Comparação
BFS
Berco_A → Patio_2 → Alfandega → Centro_Logistico

Custo:

7 + 5 + 3 = 15
Dijkstra
Berco_A → Patio_1 → Patio_2 → Patio_3 → Centro_Logistico

Custo:

4 + 2 + 2 + 4 = 12
Explicação

BFS minimiza quantidade de arestas.

Dijkstra minimiza soma dos pesos.

```python
# Exercício 12 — Análise de rotas na rede logística

import heapq
from collections import defaultdict, deque

grafo = defaultdict(list)

arestas = [
    ("Berco_A", "Patio_1", 4),
    ("Berco_A", "Patio_2", 7),
    ("Berco_B", "Patio_2", 3),
    ("Berco_B", "Patio_3", 6),
    ("Patio_1", "Patio_2", 2),
    ("Patio_2", "Patio_3", 2),
    ("Patio_1", "Alfandega", 8),
    ("Patio_2", "Alfandega", 5),
    ("Patio_3", "Centro_Logistico", 4),
    ("Alfandega", "Centro_Logistico", 3)
]

for u, v, peso in arestas:
    grafo[u].append((v, peso))
    grafo[v].append((u, peso))

# DFS
visitados = set()
ordem_dfs = []

def dfs(v):
    visitados.add(v)
    ordem_dfs.append(v)

    for vizinho, _ in grafo[v]:
        if vizinho not in visitados:
            dfs(vizinho)

dfs("Berco_A")

print("DFS:", ordem_dfs)

# BFS
fila = deque(["Berco_A"])

visitados = set(["Berco_A"])

ordem_bfs = []

while fila:
    atual = fila.popleft()

    ordem_bfs.append(atual)

    for vizinho, _ in grafo[atual]:
        if vizinho not in visitados:
            visitados.add(vizinho)
            fila.append(vizinho)

print("BFS:", ordem_bfs)

# Dijkstra
def dijkstra(origem):
    dist = {v: float("inf") for v in grafo}
    dist[origem] = 0

    heap = [(0, origem)]

    while heap:
        custo, atual = heapq.heappop(heap)

        if custo > dist[atual]:
            continue

        for vizinho, peso in grafo[atual]:
            novo = custo + peso

            if novo < dist[vizinho]:
                dist[vizinho] = novo
                heapq.heappush(heap, (novo, vizinho))

    return dist

distancias = dijkstra("Berco_A")

print("Dijkstra:", distancias)
```

Análise
DFS
Explora profundamente
Boa para análise estrutural e componentes
BFS
Explora por camadas
Ideal para menor caminho em grafos não ponderados
Dijkstra
Considera pesos
Ideal para otimização de custo
Limitações
DFS

Não garante menor caminho.

BFS

Ignora custos.

Dijkstra

Mais custoso computacionalmente que BFS/DFS.

Complexidades
DFS
O(V + E)
BFS
O(V + E)
Dijkstra
O((V + E) log V)
