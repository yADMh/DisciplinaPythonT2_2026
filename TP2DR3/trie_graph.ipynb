````python
vertices = [
    "produto",
    "processador",
    "programa",
    "python",
    "proxy",
    "printer"
]

trie = Trie()
graph = GraphAdjList()

for v in vertices:
    trie.insert(v)
    graph.add_vertex(v)

def find_vertices_by_prefix(prefix, k):
    candidates = trie.autocomplete(prefix, k)

    result = []

    for c in candidates:
        if c in graph.adj:
            result.append(c)

    return result

print(find_vertices_by_prefix("pro", 5))
print(find_vertices_by_prefix("py", 5))
print(find_vertices_by_prefix("pri", 5))
````

# Complexidade da Trie

## Tabela de complexidade

| Operação | Complexidade |
|---|---|
| insert(word) | O(n) |
| search(word) | O(n) |
| starts_with(prefix) | O(n) |
| collect_words() | O(s) |
| autocomplete(prefix, k) | O(p + s + n log n) |
| autocorrect(word) | O(p + s) |

---

## Explicação das variáveis

- n = tamanho da palavra
- p = tamanho do prefixo
- s = tamanho da subárvore percorrida

---

## Análise das operações

### insert(word)

A operação percorre todos os caracteres da palavra e cria nós quando necessário.

Complexidade:

```text
O(n)
```

---

### search(word)

A busca percorre cada caractere da palavra verificando se o caminho existe na Trie.

Complexidade:

```text
O(n)
```

---

### starts_with(prefix)

A operação percorre apenas os caracteres do prefixo.

Complexidade:

```text
O(n)
```

---

### collect_words()

Percorre recursivamente a subárvore coletando todas as palavras existentes.

Complexidade:

```text
O(s)
```

---

### autocomplete(prefix, k)

O algoritmo:
1. Localiza o prefixo.
2. Percorre a subárvore.
3. Ordena os resultados.
4. Retorna no máximo k palavras.

Complexidade aproximada:

```text
O(p + s + n log n)
```

---

### autocorrect(word)

O algoritmo procura o maior prefixo compartilhado entre a palavra buscada e as palavras armazenadas na Trie.

Complexidade aproximada:

```text
O(p + s)
```

---

# Comparação entre representações de Grafo

## Tabela comparativa

| Operação | Lista de Adjacência | Matriz de Adjacência |
|---|---|---|
| Inserção de aresta | O(1) | O(1) |
| Consulta de adjacência | O(grau do vértice) | O(1) |
| Consumo de memória | O(V + E) | O(V²) |

---

# Análise das representações

## Lista de adjacência

A lista de adjacência armazena apenas as conexões existentes entre vértices.

### Vantagens

- Menor consumo de memória
- Ideal para grafos esparsos
- Estrutura mais flexível

### Desvantagens

- Consulta de adjacência mais lenta
- Necessita percorrer vizinhos

---

## Matriz de adjacência

A matriz de adjacência utiliza uma matriz NxN contendo 0 ou 1 para representar conexões.

### Vantagens

- Consulta de adjacência em O(1)
- Implementação simples para verificar conexões

### Desvantagens

- Alto consumo de memória
- Pouco eficiente para grafos grandes e esparsos

---

# Escolha da representação em diferentes cenários

## Grafo esparso com milhões de vértices

A melhor escolha é:

```text
Lista de adjacência
```

### Justificativa

Grafos esparsos possuem poucas arestas em relação ao número de vértices. A lista de adjacência consome memória proporcional apenas às conexões existentes.

Complexidade de memória:

```text
O(V + E)
```

Isso torna a estrutura viável para grandes volumes de dados.

---

## Grafo denso com poucos vértices

A melhor escolha é:

```text
Matriz de adjacência
```

### Justificativa

Em grafos densos, quase todos os vértices possuem conexão entre si. Nesse cenário, a matriz oferece consultas extremamente rápidas.

Complexidade de consulta:

```text
O(1)
```

Como o número de vértices é pequeno, o alto consumo de memória deixa de ser um problema significativo.

# Decisões técnicas da implementação

## Decisão 1 — Uso de dicionário na Trie

Foi utilizado um dicionário (`dict`) para armazenar os filhos de cada nó.

Exemplo:

```python
self.children = {}
```

### Justificativa

O dicionário permite acesso rápido aos caracteres filhos com custo médio:

```text
O(1)
```

Isso melhora significativamente a eficiência das operações de busca e inserção.

---

## Decisão 2 — Critério de desempate no autocorrect

Quando múltiplas palavras possuem o mesmo tamanho de prefixo compartilhado, foi utilizada a menor palavra lexicograficamente.

Exemplo:

```python
suggestions.sort()
return suggestions[0]
```

### Justificativa

Essa decisão torna o comportamento determinístico e previsível, evitando resultados aleatórios.

---

## Decisão 3 — Uso de lista de adjacência no grafo principal

O grafo principal foi implementado utilizando lista de adjacência.

### Justificativa

A lista de adjacência é mais eficiente para grafos esparsos e possui menor consumo de memória.

---

## Decisão 4 — Controle de duplicação no Mermaid

Foi utilizado um conjunto (`set`) chamado `visited` para impedir duplicação de arestas na exportação Mermaid.

Exemplo:

```python
visited = set()
```

### Justificativa

Como o grafo não direcionado armazena arestas em ambos os sentidos, o conjunto evita exportar a mesma conexão duas vezes.
