```python
class BinaryHeap:
    def __init__(self):
       self.data = []
       self.operations = 0

    def parent(self, i):
        return (i - 1) // 2

    def left(self, i):
        return 2 * i + 1

    def right(self, i):
        return 2 * i + 2

    def insert(self, value):
        self.data.append(value)
        i = len(self.data) - 1

        while i > 0 and self.data[self.parent(i)] < self.data[i]:
            self.operations += 1  # conta operação
            print(f"Troca: {self.data[i]} <-> {self.data[self.parent(i)]}")
            self.data[i], self.data[self.parent(i)] = self.data[self.parent(i)], self.data[i]
            i = self.parent(i)

    def heapify(self, i):
        largest = i
        left = self.left(i)
        right = self.right(i)

        if left < len(self.data) and self.data[left] > self.data[largest]:
            largest = left

        if right < len(self.data) and self.data[right] > self.data[largest]:
            largest = right

        if largest != i:
            self.operations += 1  # conta operação
            print(f"Troca: {self.data[i]} <-> {self.data[largest]}")
            self.data[i], self.data[largest] = self.data[largest], self.data[i]
            self.heapify(largest)

    def extract_max(self):
        if not self.data:
            return None

        max_val = self.data[0]
        self.data[0] = self.data[-1]
        self.data.pop()

        self.heapify(0)
        return max_val

    def contains(self, value):
        for v in self.data:
            self.operations += 1
            if v == value:
                return True
        return False

    def delete(self, value):
        if value not in self.data:
            return False

        i = self.data.index(value)
        self.data[i] = self.data[-1]
        self.data.pop()

        self.heapify(i)
        return True

    def build_heap(self, arr):
        self.data = arr[:]
        for i in range(len(self.data)//2 - 1, -1, -1):
            self.heapify(i)
```

**Complexidade:** O(log n)

Invariantes

## Exercício 4 — Remoção do elemento máximo (extract-max)

A remoção é realizada substituindo a raiz pelo último elemento e aplicando o processo de sift-down para restaurar a propriedade da heap.

```python
print("=== EXERCÍCIO 4: EXTRACT MAX ===")

h = BinaryHeap()

# Construindo heap
h.insert(10)
h.insert(20)
h.insert(5)
h.insert(30)

print("Heap inicial:", h.data)

# Removendo elementos um por um
print("\nRemoções:")
while len(h.data) > 0:
    print("Removido:", h.extract_max())
    print("Heap agora:", h.data)
       

# O(log n)
```

## Exercício 5 — Busca em Heap

A busca é realizada de forma linear, pois a heap não mantém ordenação total dos elementos.

```python
print("=== EXERCÍCIO 5: BUSCA (CONTAINS) ===")

h = BinaryHeap()

# Inserindo valores
h.insert(10)
h.insert(20)
h.insert(5)
h.insert(30)

print("Heap atual:", h.data)

# Testes de busca
print("\nBuscas:")
print("Contém 20?", h.contains(20))
print("Contém 5?", h.contains(5))
print("Contém 100?", h.contains(100))

# O(n)
```

## Exercício 6 — Remoção de elemento arbitrário

A remoção é feita substituindo o elemento pelo último da heap e aplicando heapify para restaurar a estrutura.

```python
print("=== EXERCÍCIO 6: DELETE ===")

h = BinaryHeap()

# Inserindo valores
h.insert(10)
h.insert(20)
h.insert(5)
h.insert(30)

print("Heap inicial:", h.data)

# Deletando valor existente
print("\nDeletando 20:")
resultado = h.delete(20)
print("Sucesso?", resultado)
print("Heap após delete:", h.data)

# Deletando valor inexistente
print("\nDeletando 100:")
resultado = h.delete(100)
print("Sucesso?", resultado)
print("Heap após tentativa:", h.data)

# O(n) (buscar) + O(log n) (reorganizar)
```

## Exercício 11 — Análise Empírica

Foi feita a medição de tempo e contagem de operações durante a inserção na heap. Os testes mostram que o número de operações cresce com o tamanho da entrada.

**Complexidade:** O(log n)

```python
import time
import random

print("=== EXERCÍCIO 11: ANÁLISE EMPÍRICA ===")

h = BinaryHeap()

# Gerando dados aleatórios
dados = [random.randint(1, 1000) for _ in range(20)]

print("Dados:", dados)

start = time.time()

# Inserindo na heap
for v in dados:
    h.insert(v)

end = time.time()

print("\nHeap final:", h.data)
print("Operações realizadas:", h.operations)
print("Tempo de execução:", end - start)
```
